Q 5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a heatmap.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Set a random seed for reproducibility
np.random.seed(53)

# Generate a synthetic dataset with correlated features
n_samples = 1000
x1 = np.random.randn(n_samples)
x2 = x1 + np.random.randn(n_samples) * 0.5
x3 = x2 + np.random.randn(n_samples) * 0.2
x4 = x1 - np.random.randn(n_samples) * 0.3

# Create a DataFrame
data = pd.DataFrame({'Feature 1': x1, 'Feature 2': x2, 'Feature 3': x3, 'Feature 4': x4})

# Calculate the correlation matrix
corr_matrix = data.corr()

# Visualize the correlation matrix using a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
import numpy as np
import pandas as pd
import plotly.express as px

np.random.seed(20)
data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}

df = pd.DataFrame(data)

# Create a pivot table to aggregate sales by month and day
pivot_table = df.pivot_table(values='Sales', index='Month', columns='Day', aggfunc='mean')

# Create the heatmap using Plotly
# Use x and y arguments to provide custom labels
fig = px.imshow(pivot_table,
                 labels=dict(x="Day of Month", y="Month", color="Sales"),
                 x=pivot_table.columns,  # Use x instead of x_labels
                 y=pivot_table.index)  # Use y instead of y_labels


fig.update_layout(title='Sales Heatmap by Month and Day')
fig.show()
```

Q 2.Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the markers based on the 'sizes' and 'colors' columns.

```python
import numpy as np
import pandas as pd
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource

# Generate random data
np.random.seed(42)  # For reproducibility
n = 100  # Number of points

# Create a DataFrame with random x, y, sizes, and colors
data = {
    'x': np.random.rand(n) * 100,  # x values between 0 and 100
    'y': np.random.rand(n) * 100,  # y values between 0 and 100
    'sizes': np.random.randint(10, 50, size=n),  # Random sizes between 10 and 50
    'colors': [f'#{np.random.randint(0, 0xFFFFFF):06x}' for _ in range(n)]  # Random colors in hex format
}

df = pd.DataFrame(data)

# Create a ColumnDataSource
source = ColumnDataSource(df)

# Create a scatter plot
output_notebook()  # Use this if you're running in a Jupyter Notebook
p = figure(title="Random Scatter Plot", x_axis_label='X Axis', y_axis_label='Y Axis')

# Add scatter glyphs with varying sizes and colors
p.scatter('x', 'y', size='sizes', color='colors', source=source, alpha=0.5)

# Show the plot
show(p)
```

Q 3. Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears'] counts = [20, 25, 30, 35]

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource

# Prepare the data
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

# Create a ColumnDataSource
source = ColumnDataSource(data=dict(fruits=fruits, counts=counts))

# Create a figure
p = figure(x_range=fruits, title="Fruit Counts", toolbar_location=None, tools="")

# Create a bar chart
p.vbar(x='fruits', top='counts', width=0.9, source=source, legend_field="fruits",
       line_color='white', fill_color='green')

# Customize the chart
p.xgrid.grid_line_color = None
p.axis.minor_tick_line_color = None
p.outline_line_color = None
p.xaxis.axis_label = "Fruits"
p.yaxis.axis_label = "Counts"
p.legend.location = "top_left"
p.legend.orientation = "horizontal"

# Show the results
output_notebook()  # Use this line to display the chart in a Jupyter Notebook
show(p)
```

Q 5. Create a Bokeh heatmap using the provided dataset.
data_heatmap = np.random.rand(10, 10) x = np.linspace(0, 1, 10) y = np.linspace(0, 1, 10) xx, yy = np.meshgrid(x, y)

```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.models import ColorBar, LinearColorMapper
from bokeh.transform import transform
from bokeh.io import output_notebook

# Enable notebook output
output_notebook()

# Generate random data
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

# Prepare data for the heatmap
data = {'x': xx.flatten(), 'y': yy.flatten(), 'value': data_heatmap.flatten()}

# Create a color mapper
mapper = LinearColorMapper(palette="Viridis256", low=data['value'].min(), high=data['value'].max())

# Create a Bokeh figure
p = figure(title="Heatmap Example", x_axis_label='X', y_axis_label='Y',
           x_range=(0, 1), y_range=(0, 1), toolbar_location=None)

# Create the heatmap using rectangles
p.rect(x='x', y='y', width=0.1, height=0.1, source=data,
        fill_color=transform('value', mapper), line_color=None)

# Add color bar
color_bar = ColorBar(color_mapper=mapper, location=(0, 0))
p.add_layout(color_bar, 'right')

# Show the plot
show(p)
```
